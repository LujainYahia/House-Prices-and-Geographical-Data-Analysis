```python
import pandas as pd
pd.options.plotting.backend = 'plotly'
```

```python
df = pd.read_csv('HousePrices.csv',index_col=['date'],parse_dates=True)
df.sort_index(inplace=True)
```

```python
df.head()
```

Now, let's plot a simple line plot.

```python
df['price'].plot()
```

```python
import plotly.express as px
```

```python
fig = px.line(df,x=df.index,y = 'price',title='Line Plot using Plotly')
fig.show()
```

```python
fig = px.histogram(df,x=df['price'])
fig.show()
```

```python
fig = px.box(df,x=df['price'],title='Box Plot')
fig.show()
```

```python
dir(px)
```

```python
fig = px.scatter_3d(df,x=df.index,y=df['price'],z=df['sqft_living'],title='3D graph using plotly')
fig.show()
```
